# **1. About** <a class="anchor" id="1"></a>

A decision tree is a flowchart-like structure used in machine learning and data analysis to make decisions or predictions based on input data. It is a simple yet powerful algorithm that can be used for both classification and regression tasks.

The decision tree starts with a single node called the root node, which represents the entire dataset. The tree then splits the data based on different features or attributes, creating branches and sub-nodes. Each split is determined by a specific condition or rule based on the values of the features

At each internal node, a decision is made based on the condition associated with that node, and the data is divided into smaller subsets. This process continues recursively until a stopping criterion is met. The stopping criterion can be based on various factors such as the maximum depth of the tree, the minimum number of samples required at a leaf node, or the maximum impurity reduction.

The final nodes of the decision tree are called leaf nodes or terminal nodes, which provide the output or prediction. For classification tasks, each leaf node represents a specific class label, while for regression tasks, the leaf nodes contain continuous values representing the predicted output.

The decision tree algorithm aims to find the optimal splits that maximize the information gain or minimize the impurity measure at each step. Different impurity measures can be used, such as Gini impurity or entropy, depending on the specific problem and desired outcomes.

Once the decision tree is built, it can be used to make predictions or classify new instances by following the path from the root node to the appropriate leaf node based on the feature values of the input.

Overall, decision trees are popular due to their interpretability, as the resulting tree structure can be easily visualized and understood. They are also capable of handling both categorical and numerical data and can capture non-linear relationships between variables. However, decision trees are prone to overfitting if not properly regularized, and they may struggle with complex datasets or datasets with high dimensionality.



# **2. Purity in Decision Tree** <a class="anchor" id="2"></a>


purity refers to the homogeneity of the target variable within a subset of data. It is a measure of how well a node separates the data into distinct classes or categories.

There are different impurity measures used to assess the purity of a node, and the choice of measure depends on the specific problem and the type of target variable (categorical or continuous).

Two commonly used impurity measures for classification tasks are Gini impurity and entropy:

- Gini impurity: It measures the probability of misclassifying a randomly chosen element in a subset. The Gini impurity for a node is calculated as the sum of the squared probabilities of each class label being chosen randomly from the subset. A lower Gini impurity indicates higher purity.


- Entropy: It measures the average amount of information required to classify an element in a subset. The entropy for a node is calculated as the sum of the probabilities of each class label multiplied by their logarithms (base 2). A lower entropy indicates higher purity.



Both impurity measures aim to minimize the impurity or maximize the purity of the subsets produced by splitting the data based on different attributes. The decision tree algorithm evaluates different attributes and splits the data based on the attribute that leads to the highest reduction in impurity. This reduction is often quantified using metrics such as information gain (for entropy) or Gini gain (for Gini impurity).

The process of splitting continues recursively, with the goal of creating subsets that are as pure as possible until a stopping criterion is met. This stopping criterion could be reaching a maximum tree depth, having a minimum number of samples at a node, or exceeding a predefined impurity threshold.

By maximizing purity at each node and creating homogeneous subsets, decision trees aim to achieve accurate predictions or classifications for new instances based on the learned patterns and rules within the tree structure.

# **3.  Selection measures** <a class="anchor" id="3"></a>



Selection measures are used to evaluate and select the best attribute or feature to split the data at each node. These measures quantify the effectiveness of a split in terms of reducing impurity and increasing the homogeneity or purity of the resulting subsets.

Here are some commonly used selection measures:

- Information Gain: Information gain is a measure based on entropy and is used with categorical target variables. It calculates the reduction in entropy achieved by splitting the data based on a particular attribute. The attribute with the highest information gain is selected as the splitting attribute. Information gain is computed as the difference between the entropy of the parent node and the weighted sum of the entropies of the child nodes.

- Gini Gain: Gini gain is a measure based on Gini impurity and is also used with categorical target variables. It calculates the reduction in Gini impurity achieved by splitting the data based on a specific attribute. The attribute with the highest Gini gain is chosen as the splitting attribute. Gini gain is calculated as the difference between the Gini impurity of the parent node and the weighted sum of the Gini impurities of the child nodes.

- Gain Ratio: Gain ratio is a modification of information gain that takes into account the intrinsic information of an attribute. It overcomes the bias towards attributes with a large number of distinct values. Gain ratio is calculated by dividing the information gain by the split information, which is a measure of the potential information generated by the attribute.

- Variance Reduction: Variance reduction is a selection measure used in regression tasks, where the target variable is continuous. It quantifies the reduction in variance achieved by splitting the data based on a specific attribute. The attribute that leads to the highest reduction in variance is selected as the splitting attribute.

These selection measures help the decision tree algorithm determine the best attribute to split the data at each node, aiming to maximize the reduction in impurity or variance and create subsets that are more homogeneous or pure. The selected attribute becomes the decision rule for that node in the decision tree.

# **4. Overfitting In Decision Tree** <a class="anchor" id="4"></a>


Overfitting is a common issue that can occur in decision trees when the model becomes overly complex and starts to memorize the training data instead of generalizing well to new, unseen data. This can lead to poor performance and inaccurate predictions.

Several methods can be employed to address overfitting in decision trees:

Pre-pruning and post-pruning are two approaches to address overfitting in decision trees. Both techniques aim to prevent the tree from becoming overly complex and memorizing the training data, thereby improving the generalization ability of the model.

1. Pre-pruning:
Pre-pruning involves setting stopping criteria during the construction process of the decision tree. The tree is pruned as it grows, preventing it from expanding excessively. Some common pre-pruning techniques include:
Maximum Depth: Limiting the maximum depth of the tree by specifying the maximum number of levels. This ensures that the tree doesn't grow too deep and helps prevent overfitting.

- Minimum Sample Split: Setting a minimum number of samples required to split a node. If the number of samples at a node is below this threshold, the node will not be split further. This helps avoid creating branches based on too few instances, which may be prone to overfitting.

- Minimum Leaf Samples: Specifying a minimum number of samples required at a leaf node. If a leaf node has fewer samples than this threshold, the node is not created, and the corresponding branch is pruned. This prevents the creation of small, isolated branches based on limited data.

Pre-pruning helps to control the growth of the tree and simplifies its structure, reducing the risk of overfitting. However, setting the right values for pre-pruning parameters requires careful consideration and experimentation.

2. Post-pruning:
Post-pruning, also known as backward pruning or cost-complexity pruning, involves constructing the decision tree to its full extent and then removing branches or nodes based on pruning criteria. The tree is pruned after the initial construction, using separate validation data or statistical tests. Post-pruning techniques include:
- Cross-Validation: The decision tree is evaluated on a validation dataset using techniques like k-fold cross-validation. The nodes or branches that do not improve the performance on the validation data significantly are pruned.

- Statistical Tests: Statistical tests, such as chi-square tests, can be used to determine the significance of a node or branch in improving the overall accuracy or impurity reduction. Nodes that do not meet a certain statistical significance level are pruned.

Post-pruning allows the decision tree to grow freely during construction, capturing all possible patterns, and then selectively removes unnecessary branches. This approach helps to refine the tree structure and remove overfitting caused by noise or outliers in the training data.

Both pre-pruning and post-pruning are effective techniques to address overfitting in decision trees. Pre-pruning controls the tree's growth during construction, while post-pruning selectively prunes branches after construction based on evaluation criteria. The choice between pre-pruning and post-pruning depends on the specific problem, available data, and the desired balance between simplicity and predictive performance.

# **8. Separating Features and Target and Encoding** <a class="anchor" id="8"></a>

```python
x = df.drop(['class'], axis = 1)
y = df['class']
```

```python
label_encoder = preprocessing.LabelEncoder()
for column in x.columns:
    x[column] = label_encoder.fit_transform(x[column])
y = label_encoder.fit_transform(y)
```

# **9. Splitting Dataset to training and test data** <a class="anchor" id="9"></a>


```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=50)
```

- Splitting the dataset into 70% for train and 30% for test

## **Model - 1** <a class="anchor" id="11"></a>


```python
Tree_entropy = DecisionTreeClassifier(criterion='entropy')
Tree_entropy.fit(x_train, y_train)
```

```python
y_pred = Tree_entropy.predict(x_test)
```

```python
print('Accuracy rate is :', accuracy_score(y_test, y_pred), '\n')
print(classification_report(y_test, y_pred))
```
